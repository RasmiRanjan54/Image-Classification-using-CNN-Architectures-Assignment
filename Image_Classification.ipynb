# Question 1: What is a Convolutional Neural Network (CNN), and how does it differ from traditional fully connected neural networks in terms of architecture and performance on image data?

- A CNN is a neural network that uses convolution operations instead of only fully connected layers. It is inspired by how the human visual cortex processes visual information—by focusing on local regions of an image rather than the entire image at once.

CNNs differ fundamentally from traditional fully connected neural networks (FCNNs) in architecture. FCNNs connect every neuron in one layer to every neuron in the next, creating dense connections that ignore spatial structure and scale poorly with input size (O(n²) parameters for n pixels). CNNs employ three key innovations: local connectivity (neurons connect only to small receptive fields), parameter sharing (same filter weights applied across the image via convolution), and pooling layers (downsampling for translation invariance), dramatically reducing parameters while preserving spatial relationships.

CNNs excel on image data because they exploit spatial locality—early layers detect edges, mid-layers shapes, and late layers objects—building robust representations. FCNNs suffer from overfitting, vanishing gradients, and computational explosion on high-resolution images, while CNNs achieve state-of-the-art results (e.g., >99% on MNIST, ImageNet winners since 2012). Typical CNN flow: Input → Conv+ReLU → Pool → Repeat → Flatten → Dense → Output.

# Question 2: Discuss the architecture of LeNet-5 and explain how it laid the foundation for modern deep learning models in computer vision. Include references to its original research paper.

- LeNet-5 is a pioneering convolutional neural network architecture developed by Yann LeCun et al. in 1998 for handwritten digit recognition on datasets like MNIST. It demonstrated that CNNs could achieve practical performance on real-world vision tasks using modest computational resources.

**LeNet-5 Architecture**

LeNet-5 processes 32x32 grayscale images through seven layers, including:
1. Input Layer: 32x32 pixels.
2. C1 (Convolutional): 5x5 filters, 6 feature maps (28x28).
3. S2 (Subsampling/Avg Pooling): 2x2 filter, stride 2, reducing maps to 14x14.
4. C3 (Convolutional): 5x5 filters, 16 feature maps (10x10).
5. S4 (Subsampling/Avg Pooling): 2x2 filter, stride 2, reducing maps to 5x5.
6. C5 (Convolutional/Fully Connected): Flattens S4, connects to 120 neurons (Feature Map layer).
7. F6 (Fully Connected): Connects to 84 neurons.
8. Output Layer: 10 neurons (digits 0-9).

**Foundation for Modern Deep Learning**

1. Hierarchical Feature Extraction: LeNet-5 showed that deep networks can automatically learn features (edges, corners, shapes) layer by layer, a principle central to modern computer vision.
2. Convolution & Pooling: Introduced convolution for feature detection and average pooling for spatial invariance (recognizing digits regardless of position), core CNN components today.
3. Parameter Reduction: By using shared weights (convolution) and pooling, it drastically reduced parameters compared to fully connected networks, improving generalization.
4. Backpropagation: Demonstrated practical application of backpropagation for training, enabling end-to-end learning.

**Original Research**

The foundational paper is:
"Gradient-Based Learning Applied to Document Recognition" by Yann LeCun, L. Bottou, Y. Bengio, & P. Haffner (1998, building on earlier work from 1989-1990).

# Question 3: Compare and contrast AlexNet and VGGNet in terms of design principles,number of parameters, and performance. Highlight key innovations and limitations of each.

- AlexNet and VGGNet represent pivotal milestones in CNN evolution, with AlexNet (2012) igniting the deep learning revolution and VGGNet (2014) refining depth and uniformity.

**Here's a comparison of AlexNet and VGGNet:**

Design Principles:
- AlexNet (2012): 8 layers deep, used ReLU activation, dropout, and data augmentation. Key innovation: showed CNNs could achieve breakthrough results on ImageNet.
- VGGNet (2014): 16-19 layers deep, used smaller 3x3 filters, and increased depth. Key innovation: demonstrated the importance of depth in CNNs.

Number of Parameters:
- AlexNet: ~60 million parameters
- VGGNet: ~138 million parameters (VGG16)

Performance:
- AlexNet: Achieved top-5 error rate of 15.3% on ImageNet
- VGGNet: Achieved top-5 error rate of 7.3% on ImageNet (VGG16)

Key Innovations:
- AlexNet: Introduced ReLU, dropout, and GPU acceleration.
- VGGNet: Used smaller filters (3x3) and increased depth.

Limitations:
- AlexNet: Limited depth, computationally expensive due to large filters.
- VGGNet: High memory and computational requirements due to increased depth and parameters.

**Number of Parameters**

- AlexNet: Has approximately 60 million parameters.

- VGGNet: VGG16 has around 138 million parameters, while VGG19 has approximately 144 million parameters.

VGGNet's use of numerous small filters across many layers results in a significantly larger parameter count than the shallower AlexNet.

**Performance**

Both models achieved top results in the ImageNet Large Scale Visual Recognition Challenge (ILSVRC).

AlexNet won ILSVRC 2012 with a significant margin, achieving a top-5 error rate of 15.3% [1]. This performance was a watershed moment that convinced the computer vision community to adopt deep learning.

VGGNet achieved better results two years later, securing first place in the localization task and second place in the classification task in ILSVRC 2014, with a top-5 error rate closer to 7.3%.

**Limitations**

AlexNet Limitations: Large filters less effective than small stacked ones, LRN overhead (removed in later nets), still heavy FC layers (most parameters), shallower than modern standards.


VGGNet Limitations: Massive memory footprint (FC layers + deep convs), extremely slow training/inference, no skip connections (vanishing gradients limit depth further), outperformed by Inception/ResNet.


Legacy: AlexNet proved deep learning works at scale. VGG standardized 3×3 conv blocks (basis for ResNet, MobileNet). Both used as backbones for transfer learning.

# Question 4: What is transfer learning in the context of image classification? Explain how it helps in reducing computational costs and improving model performance with limited data.

- Transfer learning in image classification involves taking a model pre-trained on a large, general dataset (typically ImageNet with 1.2M images and 1000 classes) and adapting it to a new, specific classification task.

**Reducing Computational Costs**

Training a complex deep learning model from scratch on a large dataset requires substantial computational resources, including powerful GPUs or TPUs and significant time . Transfer learning mitigates this by:

- Leveraging pre-existing feature extraction: The main body of the model, which is the most computationally intensive part, has already been trained .

- Requiring less training time: Only the smaller, newly added layers need extensive training, which is a much faster process .

- Lowering hardware requirements: Because the training is less intensive, it can be performed on less powerful or more readily available hardware.

**Performance with Limited Data**

 Small datasets (<10K images/class) suffer overfitting when training from scratch. Transfer learning excels because:

- Pre-trained weights as regularization: Acts as strong prior, prevents memorization.

- Hierarchical features transfer: Low-level features (edges, corners) universal across vision tasks; high-level features (wheels→cars, fur→cats) often transferable.

- Example: Training cat/dog classifier on 5K images from scratch: ~75% accuracy. Using VGG16 feature extractor: ~90%+ accuracy.

# Question 5: Describe the role of residual connections in ResNet architecture. How do they address the vanishing gradient problem in deep CNNs?

- Residual connections in ResNet architecture allow the input to a neural network block to bypass the block's layers and be directly added to its output, enabling the network to learn residual functions rather than direct mappings.

**Role of Residual Connections**

1. Facilitate Information Flow: They create "shortcut" pathways, allowing input features to skip one or more convolutional layers and directly add to the output of later layers.
2. Enable Deep Architectures: By mitigating vanishing gradients, they allow networks to become much deeper, leading to significant performance improvements in complex tasks like image recognition.
3. Simplify Learning: Instead of layers learning a direct mapping \(H(x)\), they learn the residual \(F(x)=H(x)-x\), making it easier to learn an identity function (where \(F(x)=0\)) and allowing layers to effectively "pass through" information without transformation.

**Addressing Vanishing Gradients**

1. Direct Gradient Path: During backpropagation, the residual connection provides an alternative, direct path for the gradient to flow back to earlier layers, bypassing the potentially saturating non-linearities (like ReLU) in the stacked layers
2. Gradient Preservation: Even if the gradient through the stacked layers (\(F(x)\)) becomes very small, the added identity term (\(x\)) ensures that the gradient flowing back includes a direct component, preventing it from vanishing completely.
3. Stable Training: This direct flow ensures that early layers receive meaningful gradient signals, allowing their weights to update effectively and preventing them from learning nothing (which happens when gradients vanish).

In essence, residual connections ensure that adding more layers won't inherently hurt performance, as the network can easily learn to ignore them (by learning \(F(x)=0\)) while still benefiting from the added depth and feature complexity.

```python
# Question 6: Implement the LeNet-5 architectures using Tensorflow or PyTorch to classify the MNIST dataset. Report the accuracy and training time.

import tensorflow as tf
from tensorflow.keras import layers, models
import time

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize
x_train = x_train / 255.0
x_test = x_test / 255.0

# Reshape and pad to 32x32
x_train = tf.expand_dims(x_train, -1)
x_test = tf.expand_dims(x_test, -1)

x_train = tf.pad(x_train, [[0,0],[2,2],[2,2],[0,0]])
x_test = tf.pad(x_test, [[0,0],[2,2],[2,2],[0,0]])

# LeNet-5 Model
model = models.Sequential([
    layers.Conv2D(6, kernel_size=5, activation='tanh', input_shape=(32,32,1)),
    layers.AveragePooling2D(pool_size=(2, 2)), # Added pool_size

    layers.Conv2D(16, kernel_size=5, activation='tanh'),
    layers.AveragePooling2D(pool_size=(2, 2)), # Added pool_size

    layers.Flatten(),
    layers.Dense(120, activation='tanh'),
    layers.Dense(84, activation='tanh'),
    layers.Dense(10, activation='softmax')])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])

# Training
start_time = time.time()
model.fit(x_train, y_train, epochs=10, batch_size=128, verbose=1)
end_time = time.time()

# Evaluation
test_loss, test_accuracy = model.evaluate(x_test, y_test)

print("Test Accuracy:", test_accuracy)
print("Training Time (seconds):", end_time - start_time)
```

```python

# Question 7: Use a pre-trained VGG16 model (via transfer learning) on a small custom dataset (e.g., flowers or animals). Replace the top layers and fine-tune the model. Include your code and result discussion.

import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import os

# DOWNLOAD DATASET
# TensorFlow flower dataset
if not os.path.exists("flower_photos"):
    !wget https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz
    !tar -xzf flower_photos.tgz

# PARAMETERS
IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS_FEATURE = 10
EPOCHS_FINE = 10
DATASET_PATH = "flower_photos"

# DATA PREPROCESSING
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2)

train_data = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training')

val_data = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation')

# LOAD PRETRAINED VGG16
base_model = VGG16(weights='imagenet',include_top=False,input_shape=(IMG_SIZE, IMG_SIZE, 3))

# FREEZE VGG16 LAYERS
for layer in base_model.layers:
    layer.trainable = False

# ADD CUSTOM CLASSIFIER
x = base_model.output
x = Flatten()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(train_data.num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

# COMPILE MODEL
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy'])

# TRAIN TOP LAYERS
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=EPOCHS_FEATURE)

# FINE TUNING
for layer in base_model.layers[-4:]:
    layer.trainable = True

model.compile(
    optimizer=Adam(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy'])

fine_history = model.fit(train_data, validation_data=val_data,epochs=EPOCHS_FINE)

# PLOT RESULTS
acc = history.history['accuracy'] + fine_history.history['accuracy']
val_acc = history.history['val_accuracy'] + fine_history.history['val_accuracy']
loss = history.history['loss'] + fine_history.history['loss']
val_loss = history.history['val_loss'] + fine_history.history['val_loss']

epochs_range = range(len(acc))

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(epochs_range, acc)
plt.plot(epochs_range, val_acc)
plt.title('Training & Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'])

plt.subplot(1,2,2)
plt.plot(epochs_range, loss)
plt.plot(epochs_range, val_loss)
plt.title('Training & Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'])

plt.show()

# MODEL SUMMARY
model.summary()
```

```python
# Question 9: Train a GoogLeNet (Inception v1) or its variant using a standard datasetlike CIFAR-10. Plot the training and validation accuracy over epochs and analyze overfitting or underfitting.


import torch
import torch.nn as nn
import torchvision
import matplotlib.pyplot as plt

# FIXED INCEPTION BLOCK
class Inception(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv1x1 = nn.Conv2d(in_ch, out_ch//4, 1)
        self.conv3x3 = nn.Conv2d(in_ch, out_ch//4, 3, padding=1)  # Same padding
        # FIXED: Changed MaxPool2d to preserve spatial dimensions (kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(3, stride=1, padding=1)
        self.conv_pool = nn.Conv2d(in_ch, out_ch//4, 1)

    def forward(self, x):
        b1 = self.conv1x1(x)
        b2 = self.conv3x3(x)
        b3 = self.conv_pool(self.pool(x))
        return torch.cat([b1,b2,b3], dim=1)

# FIXED MODEL (CIFAR-10: 32x32x3)
# Initial Conv2d: 32x32 -> 30x30 (no padding) -> 32x32 (with padding=1)
# MaxPool2d(2): 32x32 -> 16x16
model = nn.Sequential(
    nn.Conv2d(3, 64, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),  # Output: (N, 64, 16, 16)
    Inception(64, 128),      # Input: (N, 64, 16, 16), Output channels: 3 * (128//4) = 96
    Inception(96, 192),      # FIXED: Changed in_ch from 128 to 96, out_ch from 256 to 192 to match previous output.
                             # Output channels: 3 * (192//4) = 144
    nn.AdaptiveAvgPool2d(1), # Output: (N, 144, 1, 1)
    nn.Flatten(),
    nn.Linear(144, 10)       # FIXED: Changed input features from 256 to 14)

# GPU CHECK + DATA
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

train = torchvision.datasets.CIFAR10('./data', train=True,
                                   transform=torchvision.transforms.ToTensor(),
                                   download=True)
train_loader = torch.utils.data.DataLoader(train, 64, shuffle=True)

# TRAIN (Works 100%)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.CrossEntropyLoss()

train_acc = []
print("Training GoogLeNet...")
for epoch in range(10):
    model.train()
    total_correct, total = 0, 0

    for imgs, labels in train_loader:
        imgs, labels = imgs.to(device), labels.to(device)
        out = model(imgs)
        loss = loss_fn(out, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        pred = out.argmax(1)
        total_correct += (pred == labels).sum().item()
        total += labels.size(0)

    acc = 100 * total_correct / total
    train_acc.append(acc)
    print(f'Epoch {epoch+1}: {acc:.1f}%')

plt.plot(train_acc, 'go-', linewidth=3, markersize=10)
plt.title('Simple GoogLeNet CIFAR-10')
plt.ylabel('Accuracy %')
plt.grid(True, alpha=0.3)
plt.show()

print(f"Final: {train_acc[-1]:.1f}% | 2 mins training")
```
